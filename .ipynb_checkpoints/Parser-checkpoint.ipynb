```python
import requests
from bs4 import BeautifulSoup
from fake_useragent import UserAgent
```

```python
ua=UserAgent() #Creates a fake user agent object
#We need to send user agent info in headers while sending the requests so that scrapy will not get blocked by any web pages
```

```python
header={"user-agent":ua.chrome}#Creates a fake user agent header for HTTP request which mimics the google chrome web browser
```

```python
quotes_page=requests.get("https://quotes.toscrape.com/",headers=header) #Paste the url of your interset
```

```python
quotes_page
```

```python
soup_lxml=BeautifulSoup(quotes_page.content,"lxml") #Using lxml parser to parse the content
#content of quotes_page is parsed
```

```python
print(soup_lxml.prettify())#Prettify is used to print the content with proper indentations
```

# html5lib:

#### html5lib is an HTML5-compliant, pure Python library that parses HTML documents in a manner similar to a web browser. It creates a parse tree that closely resembles how a browser would interpret the document, making it very lenient with malformed HTML.

#### Pros:
###### Excellent at handling badly formed or messy HTML.
###### Relatively easy to use.
###### Produces a parse tree that mirrors the structure of the document.

#### Cons:
##### Slower than lxml 

# html.parser 

#### html.parser is a built-in HTML parser in Python's standard library. It is based on the html.parser module and is written in Python.

#### Pros:
###### Available in the Python standard library, so no external installation is required.
###### Faster than html5lib because it's implemented in CPython.
###### Suitable for parsing well-formed HTML.

#### Cons:
##### Less tolerant of malformed HTML compared to html5lib.
##### Not as feature-rich as lxml.

# lxml:

#### lxml is a high-performance, third-party library for parsing both XML and HTML documents. It uses C libraries to speed up parsing and tree manipulation.

#### Pros:
##### Extremely fast and efficient, especially for large documents.
##### Supports both XPath and CSS selectors for navigating and querying the document.
##### Well-suited for parsing well-formed and valid HTML/XML.

#### Cons:
##### Requires external installation (pip install lxml).
##### Less tolerant of malformed HTML compared to html5lib.
