```python
import requests
from bs4 import BeautifulSoup
import csv
url = 'https://www.worldometers.info/coronavirus/countries-where-coronavirus-has-spread/'

# get URL's html
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
data = []

data_iterator = iter(soup.find_all('td'))
# data_iterator is the iterator of the table
data_iterator
```

```python
# This loop will keep repeating till there is data available in the iterator
while True:
    try:
        data_dict={}
        data_dict['country'] = next(data_iterator).text
        data_dict['confirmed'] = int(next(data_iterator).text.replace(',',''))
        data_dict['deaths'] = int(next(data_iterator).text.replace(',',''))
        data_dict['continent'] = next(data_iterator).text

        data.append(data_dict)

# StopIteration error is raised when there are no more elements left to iterate through
    except StopIteration:
        break
data
```

```python
filename = 'covid_cases.csv'
with open(filename, 'w', newline='') as f:
	w = csv.DictWriter(f,['country','confirmed','deaths','continent'])
	w.writeheader()
	for row in data:
		w.writerow(row)
```
